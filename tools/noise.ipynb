```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def prng(x: np.ubyte, y: np.ubyte) -> np.ubyte:
    x ^= y << 1
    y ^= x << 3
    return x + y
```

```python
terrain = np.empty((256, 256), dtype=np.ubyte)

for y, x in np.ndindex(terrain.shape):
    terrain[x, y] = prng(x, y)

plt.hist(np.ravel(terrain))
plt.show()
```

```python
plt.imshow(terrain, cmap='binary_r')
plt.colorbar()
plt.show()
```
